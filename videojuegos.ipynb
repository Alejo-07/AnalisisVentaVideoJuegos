```python
nuevos_nombres = {
    'Ventas NA': 'Ventas Norte America',
    'Ventas EU': 'Ventas Europa',
    'Ventas JP': 'Ventas Japon',
    'Ventas Otros': 'Ventas Otras Regiones'
}

# Cambiar los nombres de las columnas usando el método 'rename'
df = df.rename(columns=nuevos_nombres)
```

```python
df.shape
```

```python
def categorizar_generacion(año):
    if 1970 <= año < 1980:
        return 'Década de los 70'
    elif 1980 <= año < 1990:
        return 'Década de los 80'
    elif 1990 <= año < 2000:
        return 'Década de los 90'
    elif 2000 <= año < 2010:
        return 'Década de los 2000'
    elif 2010 <= año < 2020:
        return 'Década de los 2010'
    else:
        return 'Otra'

# Aplicar la función a la columna 'Año' y crear la nueva columna 'Generacion'
df['Generacion'] = df['Año'].apply(categorizar_generacion)
```

```python
df['Año'] = df['Año'].astype(str)
```
